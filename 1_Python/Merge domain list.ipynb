```python
import csv
import os
import pandas as pd


directory = '../domains/sublists'



domain_list = []


    
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    try: 
        df_item = pd.read_csv(file_path, sep=';')
        for index, row in df_item.iterrows():
            item_dict = {'domain': row['domain'], 'source' : [row['source']]}
            if any(x['domain'] == row['domain'] for x in domain_list):
                for x in domain_list:
                    if x['domain'] == row['domain']:
                        x['source'].append(row['source'])
            else:
                domain_list.append(item_dict)
    except: 
        df_item = pd.read_csv(file_path, sep=',')
        for index, row in df_item.iterrows():
            item_dict = {'domain': row['domain'], 'source' : [row['source']]}
            if any(x['domain'] == row['domain'] for x in domain_list):
                item_dict['source'].append(row['source'])
            else:
                domain_list.append(item_dict)
                
print(domain_list)

df = pd.DataFrame(domain_list)

df.to_csv('../domains/Domains_and_Sources.csv')
        
```
